```python
import json
import numpy as np
import pandas as pd
```

```python
def read_data(name: str):
    with open(name, "r+") as f:
        f.seek(0)
        data = json.load(f)
    return data
```

```python
def save_data(name: str, data: {}):
    with open(name, "w+") as f:
        f.seek(0)
        json.dump(data, f)
        f.truncate()
```

```python
final = {}
```

```python
ackley = read_data("results/ackley.json")

for test_name, test_dims in ackley.items():
    final[test_name] = {}

    for test_dim, tabu_lens in test_dims.items():
        final[test_name][test_dim] = {
            "constant": {"mean": 10001},
            "logarithmic": {"mean": 10001},
            "geometric": {"mean": 10001},
            "adaptive": {"mean": 10001}
        }

        for tabu_len, sched_names in tabu_lens.items():
            for sched_name, sched_params in sched_names.items():
                best = {"mean": 10001}

                for param_set, param_stats in sched_params.items():
                    if param_stats["mean"] < best["mean"]:
                        best = param_stats
                        best["params"] = param_set
                        best["tabu"] = tabu_len

                if best["mean"] < final[test_name][test_dim][sched_name]["mean"]:
                    final[test_name][test_dim][sched_name] = best
```

```python
save_data("results/final.json", final)
```

```python
df = []

for test_name, test_dims in final.items():
    for test_dim, sched_names in test_dims.items():
        for sched_name, sched_info in sched_names.items():
            df.append((test_name, test_dim, sched_name, sched_info["tabu"], sched_info["params"], sched_info["min"], sched_info["max"], sched_info["mean"], sched_info["std"]))
```

```python
df = pd.DataFrame(df, columns=["problem", "dimension", "cooling schedule", "tabu length", "parameters", "min", "max", "mean", "std"])
```
